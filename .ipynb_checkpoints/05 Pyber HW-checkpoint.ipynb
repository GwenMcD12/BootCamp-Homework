```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import
city_data = ("Boot Camp Homework/HOMEWORK 05/Instructions/Pyber/raw_data/city_data.csv")
ride_data = ("Boot Camp Homework/HOMEWORK 05/Instructions/Pyber/raw_data/ride_data.csv")

city_data_df = pd.read_csv(city_data)
ride_data_df = pd.read_csv(ride_data)
```

```python
pyber_df = pd.merge(city_data_df, ride_data_df, on="city", how='left')
pyber_df.head()
```

```python
#Average Fare ($) Per City
avg_fare = pyber_df.groupby('city')['fare'].mean().to_frame()

#Total Number of Rides Per City
city_total_rides = pyber_df.groupby('city')['ride_id'].count().to_frame()

#Total Number of Drivers Per City
city_driver_count = pyber_df.groupby('city')['driver_count'].unique().to_frame()
city_driver_count['driver_count'] = city_driver_count['driver_count'].str.get(0)

#City Type (Urban, Suburban, Rural)
city_type = pyber_df.groupby('city')['type'].unique().to_frame()
city_type['type'] = city_type['type'].str.get(0)


test = avg_fare.join(city_total_rides, how = 'outer', on = 'city')
test2 = city_driver_count.join(city_type, how = 'outer', on='city')
test3 = test.join(test2, how='outer', on='city')
test3 = test3.reset_index()
test3 = test3.set_index(['type'])

test3.head(10)

I_hate_u = test3.loc[['Urban'], ['city', 'driver_count', 'fare']]
I_hate_u.head(100)
#I COPIED THE SECOND LINE OF CODE FOR CITY_TYPE AND CITY_DRIVER, I AM NOT SURE WHY I NEED IT BUT I KNOW IT DOES NOT WORK WITHOUT IT

```
